```python
import os
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set_theme()
%matplotlib inline
```

```python
d=pd.read_table('data.txt',index_col=0)
d.head()
```

```python
print(d.shape)
d.describe()
```

```python
response=d['Class'].value_counts()
print("Proportion of Negative Class:",response[0]/sum(response))
print("Proportion of Positive Class:",response[1]/sum(response))
response.plot(kind='bar')
```

### Standardize features

```python
from sklearn.preprocessing import StandardScaler
y=d.iloc[:,-1]
features=StandardScaler().fit_transform(d.iloc[:,:-1])
X = pd.DataFrame(features, index=d.index, columns=d.columns[:-1])
X.head()
```

### Heatmap of Standardized features:

```python
ax = plt.axes()
sns.heatmap(X, ax = ax)
ax.set_title('Standardized features')
plt.show()
```

```python
# 5-folds Cross Validation
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.metrics import confusion_matrix
clf = RandomForestClassifier(n_estimators=50,criterion='gini',max_depth=5,min_samples_split=2,min_samples_leaf=10,max_features='sqrt',bootstrap=True,
    n_jobs=-1,random_state=0,class_weight='balanced')
y_predict=[]
y_true=[]
for (train_id, test_id) in cv.split(X, y):
    Xtrain=X.iloc[train_id]
    ytrain=y.iloc[train_id]
    Xtest=X.iloc[test_id]
    ytest=y.iloc[test_id]
    clf.fit(Xtrain,ytrain)
    ypred=clf.predict_proba(Xtest)[:,1]
    y_predict=np.concatenate([y_predict,ypred])
    y_true=np.concatenate([y_true,ytest])

# Confusion Matrix
y_predict1=(y_predict>0.5)
cm = confusion_matrix(y_true, y_predict1)
ax = plt.axes()
sns.heatmap(cm,annot=True, fmt='g', ax = ax)
ax.set_title('Confusion matrix: Random Forests')
plt.show()

from sklearn.metrics import classification_report
print(classification_report(y_true,y_predict1))
```

```python
fpr, tpr, thresholds = roc_curve(y_true, y_predict)
auc_score = auc(fpr, tpr)
print("AUROC:",auc_score)

precision, recall, thresholds = precision_recall_curve(y_true, y_predict)
average_precision = average_precision_score(y_true, y_predict)
print("AUPR:",average_precision)

fig, ax = plt.subplots(1,2,figsize=(10,5)) 
#plt.figure(figsize=(5,5))
ax[0].plot(recall, precision, linestyle='-', lw=2, color='b')
ax[0].set_xlim(0, 1)  
ax[0].set_ylim(0, 1)  
ax[0].set_title("PR curve AUPR="+"{0:0.2f}".format(average_precision), fontsize = 10)
ax[0].set_xlabel('Recall')
ax[0].set_ylabel('Precision')
#plt.legend()
#plt.show()
# plt.savefig('PRcurve_'+cell+'_960.pdf')
    
ax[1].plot(fpr,tpr,  linestyle='-', lw=2, color='b')
ax[1].set_xlim(0, 1)  
ax[1].set_ylim(0, 1)   
ax[1].set_title("ROC curve AUROC="+"{0:0.2f}".format(auc_score), fontsize = 10)
ax[1].set_xlabel('False Positive Rate')
ax[1].set_ylabel('True Positive Rate')
ax[1].plot([0, 1], [0, 1], linestyle='--', lw=2, color='r')
#plt.legend()
#plt.show()
```

```python
# ANN model:
import tensorflow as tf
#from keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier = tf.keras.models.Sequential()
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = X.shape[1]))

# Adding the second hidden layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN 
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
X_train=X
y_train=y

y_predictNN=np.array([])
y_trueNN=np.array([])
for (train_id, test_id) in cv.split(X, y):
    Xtrain,ytrain=X.iloc[train_id],y.iloc[train_id]
    Xtest,ytest=X.iloc[test_id],y.iloc[test_id]
    classifier.fit(Xtrain, ytrain, batch_size = 10, epochs = 100,verbose = 0)
    ypred=classifier.predict(Xtest)
    score, acc = classifier.evaluate(Xtrain, ytrain,batch_size=10)
    print('Train score:', score)
    print('Train accuracy:', acc)
    y_predictNN=np.append(y_predictNN,ypred)
    y_trueNN=np.append(y_trueNN,ytest)
```

```python
fpr_NN, tpr_NN, thresholds = roc_curve(y_trueNN, y_predictNN)
auc_score_NN = auc(fpr_NN, tpr_NN)
print("ANN AUROC:",auc_score)

precision_NN, recall_NN, thresholds = precision_recall_curve(y_trueNN, y_predictNN)
average_precision_NN = average_precision_score(y_trueNN, y_predictNN)
print("ANN AUPR:",average_precision)

# Confusion Matrix
from sklearn.metrics import confusion_matrix
y_predictNN1=(y_predictNN>0.5)
cm1 = confusion_matrix(y_trueNN, y_predictNN1)
ax = plt.axes()
sns.heatmap(cm1,annot=True, fmt='g', ax = ax)
ax.set_title('Confusion matrix: Neural Networks')
plt.show()

from sklearn.metrics import classification_report
print(classification_report(y_trueNN,y_predictNN1))
```
